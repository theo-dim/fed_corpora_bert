## Check some Minutes

```python
article_socket = urlopen('https://www.federalreserve.gov/fomc/minutes/20010131.htm')
article_org = BeautifulSoup(article_socket, 'html.parser')
if str(article_org).count('</p>') < 10:
    article = BeautifulSoup(str(article_org).replace('<p>', '</p><p>'), 'html.parser')
else:
    article = article_org
paragraphs = article.findAll('p')
print(len(paragraphs))
```

```python
for para in paragraphs:
    article = para.get_text()
    input_words = re.findall(r'\b([a-zA-Z]+n\'t|[a-zA-Z]+\'s|[a-zA-Z]+)\b', article.lower())
    print(len(input_words))
    print(input_words)
    print("--------------------------------------------------")
```

```python
section = -2
mom_article = []
for paragraph in paragraphs:
    if paragraph.select('strong'):
        section += 1
        if ("voting" or "vote") in paragraph.select('strong')[0].get_text().lower():
            if len(mom_article) > 0: del mom_article[-1]
            break
        mom_article.append("")
    if section >= 0:
        mom_article[section] += '\n\n' + paragraph.get_text().strip()
```

```python
pdf_file_parsed = parser.from_file('../data/FOMC/script_pdf/FOMC_PresConfScript_2019-1-30.pdf')
pdf_file_parsed['content']
```

```python
paragraphs = re.sub('(\n)(\n)+', '\n', pdf_file_parsed['content'].strip())
# paragraphs = re.sub(r'Page[ ]*\d+[ ]*[of]*[ ]*\d+', paragraphs)
paragraphs = paragraphs.split('\n')
```

```python
len(paragraphs)
```

```python
for article in paragraphs:
    input_words = re.findall(r'\b([a-zA-Z]+n\'t|[a-zA-Z]+\'s|[a-zA-Z]+)\b', article)
    print(len(input_words))
    print(input_words)
    print("--------------------------------------------------")
```

```python
for article in paragraphs:
    if not re.search(r'^(january|february|march|april|may|june|july|august|september|october|november|december|jan|feb|mar|apr|may|jun|jul|aug|sep|oct|nov|dec)', article.lower()):
        print(re.findall(r'^(january|jan)', article.lower()))
```

```python
section = -1
article_sections = []
for article in paragraphs:
    if len(re.findall(r'[A-Z]', article[:10])) > 5:
        section += 1
        article_sections.append("")
    if section >= 0:
        if not re.search('^(january|february|march|april|may|june|july|august|september|october|november|december|jan|feb|mar|apr|may|jun|jul|aug|sep|oct|nov|dec)', article.lower()):
            if not re.search('^page', article.lower()):
                article_sections[section] += '\n\n' + article

for section in article_sections: 
    input_words = re.findall(r'\b([a-zA-Z]+n\'t|[a-zA-Z]+\'s|[a-zA-Z]+)\b', section)
    print(len(input_words))
    print(section)
    print("--------------------------------------------------")
```
